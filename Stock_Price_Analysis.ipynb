```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
days = np.arange(1,31)
stock_prices = np.random.randint(100,200, size=30)

print("Stock Prices(30 days):\n", stock_prices)

avg_price = np.mean(stock_prices)
max_price = np.max(stock_prices)
min_price = np.min(stock_prices)
best_day = np.argmax(stock_prices)
worst_day = np.argmin(stock_prices)
moving_avg = np.convolve(stock_prices, np.ones(5)/5, mode='valid')

print("\nAverage Price:",avg_price)
print("Highest Price:", max_price,"on Day", best_day)
print("Lowest Price:", min_price,"on Day", worst_day)
```

```python
plt.figure(figsize=(10,6))
plt.plot(days, stock_prices, marker="o", color="blue", label="Stock Price")
plt.plot(days[4: ], moving_avg, color="red", linestyle="--", linewidth=2, label="5-day Moving Avg")
plt.title("Stock Price Trend(30 Days)")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(days, stock_prices, color="skyblue", edgecolor="black")
plt.title("Daily Stock Prices")
plt.xlabel("Days")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(stock_prices, bins=8, color="orange", edgecolor="black")
plt.title("Distribution of stock prices")
plt.xlabel("Price Range")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
