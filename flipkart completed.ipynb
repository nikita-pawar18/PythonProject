```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
import time
!pip install textblob
from textblob import TextBlob
from nltk import sent_tokenize
import requests
!pip install nltk
import nltk

```

```python
name =[]
rating =[]
review =[]
url="""https://www.flipkart.com/apple-iphone-15-black-128-gb/product-reviews/itm6ac6485515ae4?pid=MOBGTAGPTB3VS24W&lid=LSTMOBGTAGPTB3VS24WKFODHL&marketplace=FLIPKART"""

for i in range(1,6):
   new_url = url + "&page="+str(i) 
   r = requests.get(new_url)
   soup = BeautifulSoup(r.text, "html.parser")
   Names = soup.find_all("p",{"class":"_2NsDsF AwS1CA"})
   for i in Names:
     name.append(i.text)
   name 

   Ratings = soup.find_all("div", {"class":"XQDdHH Ga3i8K"})
   for i in Ratings:
    rating.append(i.text)
   rating


   Reviews = soup.find_all("div", {"class":"ZmyHeo"})
   for i in Reviews:
    review.append(i.text)
   review  
print(len(name), len(rating), len(review))  
```

```python
df = pd.DataFrame({"Customer Names":name, "Rating":rating, "Reviews":review})
df
```

```python
df["Customer Names"] = df["Customer Names"].str.title()
df
```

```python
df["Reviews"]= df["Reviews"].str.replace("READ MORE", " ")
df
```

```python
import nltk
nltk.download('punkt_tab')
  

df['Reviews_1'] = df["Reviews"].apply(sent_tokenize)
df
```

```python
from textblob import blob
def get_polarity(sentences):
    return [TextBlob(sentence).sentiment.polarity for sentence in sentences]

df['Polarity'] = df['Reviews_1'].apply(get_polarity)

df
```

```python
from statistics import mean
def calculate_average_polarity(polarities):
    return mean(polarities) if polarities else 0

df['Average_Polarity'] = df['Polarity'].apply(calculate_average_polarity)

df.head(10)
```

```python
def sentiment(pol):
    if pol >= 0.75:
        return "Extremely Positive"
    elif pol> 0:
        return "Positive"
    elif pol == 0:
        return "Neutral"
    elif pol <=-0.75:
        return "Negative"
    else:
        return "Extremely Negative"

df["sentiment"] = df["Average_Polarity"].apply(sentiment)
df
    
```

```python
df["Average_Polarity"].mean()
#the reviews are positive overall for i phone 15 128GB
```

```python
ax = sns.countplot(x=df["sentiment"])
ax.bar_label(ax.containers[0])
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

positive_reviews = df[df['sentiment'] == 'Positive']['Reviews'].dropna()
negative_reviews = df[df['sentiment'] == 'Negative']['Reviews'].dropna()

positive_text = ' '.join(positive_reviews)
negative_text = ' '.join(negative_reviews)

if positive_text.strip():
    wordcloud_pos = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud_pos, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud - Positive Reviews')
    plt.show()
else:
    print("No positive reviews found to generate word cloud.")

if negative_text.strip():
    wordcloud_neg = WordCloud(width=800, height=400, background_color='black').generate(negative_text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud_neg, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud - Negative Reviews')
    plt.show()
else:
    print("No negative reviews found to generate word cloud.")
```
