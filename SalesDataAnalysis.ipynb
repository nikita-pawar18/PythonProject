```python
import pandas as pd
import numpy as np


products = ["Laptop","Mobile","Headphones","Keyboard","Monitor"]
months = ["Jan","Feb","Mar","Apr","May","Jun"]

np.random.seed(123)


data = {
    "Month": np.random.choice(months, 30),
    "Product": np.random.choice(products,30),
    "Units_Sold": np.random.randint(10,100,30),
    "Price": np.random.randint(2000,50000,30)

}

df= pd.DataFrame(data)
print(df.head())
```

```python
df["Revenue"] = df["Units_Sold"]*df["Price"]

sales_by_product = df.groupby("Product")["Revenue"].sum()
print("\nTotal Sales by Product:\n", sales_by_product)

avg_monthly_sales = df.groupby("Month")["Revenue"].mean()
print("\nAverage Monthly Sales:\n", avg_monthly_sales)

top_product = sales_by_product.idxmax()
print("\nBest Selling Product:", top_product)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sales_by_product.plot(kind="bar",color="lightgreen",edgecolor="black")
plt.title("Total Sales by Product")
plt.ylabel("Revenue (INR)")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
avg_monthly_sales.plot(kind="line",marker="o",color="blue")
plt.title("Average Monthly Sales")
plt.ylabel("Revenue(INR)")
plt.xlabel("Month")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["Units_Sold"],bins=8, color="purple",edgecolor="black")
plt.title("Distribution of Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Frequency")
plt.show()
```
