```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
patients = ["P1","P2","P3","P4","P5"]
days = np.arange(1,8)

heart_rate = np.random.randint(60,110,size=(len(patients), len(days)))
temperature = np.random.uniform(36.0,39.0, size=(len(patients),len(days))).round(1) 
spo2 = np.random.randint(90,100, size=(len(patients), len(days)))


print("Heart Rate:\n", heart_rate)
print("\nTemperature:\n", temperature)
print("\nSpo2:\n", spo2)
```

```python
avg_hr = np.mean(heart_rate, axis=1)
avg_temp = np.mean(temperature, axis=1)
avg_spo2 = np.mean(spo2, axis=1)

critical_patients = [patients[i] for i in range(len(patients))
                     if avg_hr[i] > 100 or avg_temp[i] > 38 or avg_spo2[i] < 92]


for i, patient in enumerate(patients):
    print(f"\n {patients}:")
    print("   Avg Heart Rate :", round(avg_hr[i],1))
    print("   Avg Temp :", round(avg_temp[i],1))
    print("   Avg Temp :", round(avg_spo2[i],1))

print("\n critical patients:", critical_patients)
    
```

```python
plt.figure(figsize=(10,5))
for i, p in enumerate(patients):
    plt.plot(days, heart_rate[i], marker="o", label=p)
plt.title("Heart Rate Trends Over 7 Days")
plt.xlabel("Day")
plt.ylabel("Heart Rate (bpm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
for i,p in enumerate(patients):
    plt.plot(days, temperature[i], marker="o", label=p)

plt.title("Temperature Trends Over 7 Days")
plt.xlabel("Day")
plt.ylabel("Temperatutre")
plt.legend()
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(8,5))
plt.bar(patients, avg_spo2, color="green", edgecolor="black")
plt.title("Average SpO2 per Patient")
plt.ylabel("SpO2 (%)")
plt.ylim(85, 100)
plt.show()
```

```python
plt.figure(figsize=(8,5))
for i,p in enumerate(patients):
    plt.scatter(temperature[i], heart_rate[i], label=p, s=80)

plt.title("Temperature vs Heart Rate")
plt.xlabel("Temperature")
plt.ylabel("Heart Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(heart_rate.T, labels=patients)  
plt.title("Heart Rate Distribution")
plt.ylabel("Heart Rate")
plt.show()
```
