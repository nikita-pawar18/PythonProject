#bar plot

```python
import matplotlib.pyplot as plt
x = ["Python","C","C++","Java"]
y = [85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["Python","C","C++","Java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("Wscube",fontsize = 15)
c = ["y","b","m","g"]
plt.bar(x,y,width = 0.4,color=c,align="center")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["Python","C","C++","Java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("Wscube",fontsize = 15)
c = ["y","b","m","g"]
plt.bar(x,y,width = 0.4,color=c,edgecolor="r",linewidth=2,linestyle=":",alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["Python","C","C++","Java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("Wscube",fontsize = 15)
c = ["y","b","m","g"]
plt.bar(x,y,width = 0.4,color="g",edgecolor="r",label="Wscube")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)

plt.barh(p,y,color="r",label="popularity")
plt.barh(p,z,color="y",label="popularity1")


plt.legend()
plt.show()
```

#Scatter plot

```python
import matplotlib.pyplot as plt
x = ([1,2,3,4,5,6,7])
y = ([2,3,1,4,5,3,6])
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no =  [2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("Scatter plot",fontsize = 15)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no =  [2,3,1,4,5,3,6]
color = ["r","y","g","b","r","g","r"]
sizes = [400,200,300,700,100,500,300]
plt.scatter(day,no,c=color,s=sizes,alpha=0.5,marker = "*")
plt.title("Scatter plot",fontsize = 15)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no =  [2,3,1,4,5,3,6]
color = ["y","b","g","b","y","g","b"]
sizes = [400,200,300,700,100,500,300]
plt.scatter(day,no,c=color,s=sizes,cmap="virids")
t=plt.colorbar()
t.set_label("Colorbar",fontsize=15)
plt.title("Scatter plot",fontsize = 15)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```

#Fill between plot 

```python
import matplotlib.pyplot as plt 

x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,color="r")
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4),alpha=0.3)
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

#Working with text in matplotlib

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title("wscube",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,5,"Python")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x, y)
plt.title("wscube", fontsize=15)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Python", fontsize=15)

plt.text(2, 5, "Java", fontsize=15, style="italic", bbox={"facecolor": "y"})


plt.annotate("Python", xy=(2, 1), xytext=(3, 3), 
             arrowprops=dict(facecolor="black", shrink=0.05))

plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x, y)
plt.title("wscube", fontsize=15)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Python", fontsize=15)

plt.legend(["up"],loc=9,facecolor="red",edgecolor="yellow",framealpha=0.5,shadow=True)

plt.show()

```
