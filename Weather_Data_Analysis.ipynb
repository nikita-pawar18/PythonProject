```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
temperature = np.array([32,34,30,29,35,36,33])

print("Temperature of the week:", temperature)

avg_temp = np.mean(temperature)
max_temp = np.max(temperature)
min_temp = np.min(temperature)
hottest_day = days[np.argmax(temperature)]
coldest_day = days[np.argmin(temperature)]

print("\nAverage Temperature:",avg_temp)
print("Highest Temperature:", max_temp,"on",hottest_day)
print("Lowest Temperature:", min_temp,"on",coldest_day)

plt.figure(figsize=(8,5))
plt.plot(days,temperature,marker="o",color="blue",linewidth=2,label="Temperature")
plt.axhline(y=avg_temp,color="red",linestyle="--",label="Average Temp")
plt.title("weekly temperature variation")
plt.xlabel("Days of week:")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(days,temperature,color="Orange",edgecolor="black")
plt.title("Temperature per Day")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```
