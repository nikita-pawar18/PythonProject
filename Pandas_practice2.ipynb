#merging and concat

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

var2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.merge(var1,var2,on="A")
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var1,var2,how="right")
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var1,var2,how="left")
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var1,var2,how="outer",indicator=True)
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
import pandas as pd 

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.merge(var1,var2,left_index = True,right_index=True,suffixes=("name","id"))
```

#Group By

```python
import pandas as pd

var = pd.DataFrame({"Name":["a","b","c","d","a","b","a","c","c","b","o"],
                    "s_1":[12,13,14,12,13,14,15,23,25,16,10],
                    "s_2":[23,24,25,26,27,28,29,30,25,34,35]})

var
```

```python
var_new = var.groupby("Name")

var_new
```

```python
for x,y in var_new:
    print(x)
    print(y)
```

```python
var_new.get_group("a")
```

```python
var_new.get_group("d")
```

```python
var_new.min()
```

```python
var_new.max()
```

```python
var_new.mean()
```

```python
li = list(var_new)
li
```

#join and Append

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"C":[10,20,30,40],"D":[11,22,33,44]})

var1.join(var2)
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"C":[10,20],"D":[11,22]})

var1.join(var2)
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var2 = pd.DataFrame({"C":[10,20],"D":[11,22]})

var2.join(var1)
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var2 = pd.DataFrame({"C":[10,20],"D":[11,22]})


var2.join(var1,how="right")
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])

var2 = pd.DataFrame({"C":[10,20],"D":[11,12]},index=["a","b"])

var2.join(var1,how="inner")
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])

var2 = pd.DataFrame({"C":[10,20],"B":[11,22]},index=["a","b"])

var2.join(var1,how="inner",lsuffix="_12")
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])

var2 = pd.DataFrame({"C":[10,20],"B":[11,22]},index=["a","b"])

var2.join(var1,how="inner",rsuffix="_12")
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])

var2 = pd.DataFrame({"C":[10,20],"B":[11,22]},index=["a","b"])

var2.join(var1,how="inner",lsuffix="_12",rsuffix="_12")
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"C":[10,20],"B":[11,22]})

result = pd.concat([var1, var2], ignore_index=True)
print(result)
```

#Pivot Table and Melt Function

```python
import pandas as pd

var = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var
```

```python
pd.melt(var)
```

```python
pd.melt(var,id_vars=["eng"])
```

```python
pd.melt(var,id_vars=["days"],var_name="python")
```

```python
pd.melt(var,id_vars=["days"],var_name="python",value_name="WsCube")
```

```python
import pandas as pd

var = pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var
```

```python
var.pivot(index="days",columns="st_name")
```

```python
var.pivot_table(index="st_name",columns="days",aggfunc="mean")
```

```python
var.pivot_table(index="st_name",columns="days",aggfunc="sum")
```

```python
var.pivot_table(index="st_name",columns="days",aggfunc="sum",margins="True")
```
