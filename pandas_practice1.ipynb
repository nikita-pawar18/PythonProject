```python
import pandas as pd

x = [3,4,5,6]
var = pd.Series(x, index= ['a','s','d','f'])
print(var)
```

#Arithmetic_operations

```python
import pandas as pd

var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
print(var)
```

```python
var["C"] = var["A"]+var["B"]
print(var["C"])
```

```python
var["C"] = var["B"]-var["A"]
print(var["C"])
```

```python
var["C"] = var["A"]/var["B"]
print(var["C"])
```

```python
var1 = pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})

print(var1)
```

```python
var1["Python"] = var1["A"]<=20

var1["Python"] = var1["B"]>=16

print(var1)
```

#dropna and fillna

```python
import pandas as pd 

var = pd.read_csv("Downloads/IPL_BallByBall2008_2024(Updated).csv")
var
```

```python
var.dropna()
```

```python
var.dropna(axis=1)
```

```python
var.dropna(how="any")
```

```python
var.dropna(how = "all")
```

```python
var.dropna(subset=['Bowler'])
```

```python
var.dropna(thresh=1)
```

```python
var.fillna("Python")
```

```python
var.fillna({"Player Out":"Python","wicket_type":"C","type of extras":"Java","fielders_involved":"C++"})
```

```python
var.fillna(method="ffill")
```

```python
var.fillna(method="bfill")
```

```python
var.fillna(method="bfill",axis=1)
```

```python
var.fillna(method="ffill",axis=1)
```
